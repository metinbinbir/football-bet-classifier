```python
import warnings
warnings.filterwarnings("ignore")

import sklearn
import pickle
import pandas as pd
from sklearn import preprocessing
```

```python
from IPython.display import display, HTML
```

```python
custom_test_data = pd.read_csv("custom_test.csv",encoding="utf-8", index_col=None)
best_ms_model = pickle.load(open("MS_numerik_(Original Data)_ _0.5615942028985508.sav", 'rb'))
best_iy_model = pickle.load(open("IY_numerik_(Original Data)_ _0.5036231884057971.sav", 'rb'))
best_kg_model = pickle.load(open("KG_numerik_(Original Data)_ _0.6141304347826086.sav", 'rb'))
```

```python
display(HTML(custom_test_data.to_html()))
```

```python
model_list = [best_ms_model, best_iy_model, best_kg_model]
```

```python
id_input = int(input("MS, İY ve KG sonucu öğrenmek için yukarıdaki tablodan en soldaki idlerden numara seçiniz (0-39 arası)"))
selected_row = custom_test_data.iloc[id_input]
    
ms_model = model_list[0]
iy_model = model_list[1]
kg_model = model_list[2]

ms_true = selected_row['MS_numerik']
ms_score = selected_row['MS']

iy_true = selected_row['IY_numerik']
iy_score = selected_row['IY']

splitted_ms = ms_score.split("-")
if int(splitted_ms[0]) > 0 and int(splitted_ms[1]) > 0:
    kg_true = 1
else:
    kg_true = 0

selected_row_true = [ms_true, iy_true, kg_true]

ms_test_input = selected_row.drop(['MS','IY','MS_numerik']).tolist()
ms_test_input.pop(0)

iy_test_input = selected_row.drop(['MS','IY','IY_numerik']).tolist()
iy_test_input.pop(0)

kg_test_input = selected_row.drop(['MS','IY','KG_numerik']).tolist()
kg_test_input.pop(0)

ms_res = ms_model.predict([ms_test_input])
iy_res = iy_model.predict([iy_test_input])
kg_res = kg_model.predict([kg_test_input])

print("---------------------------------------------------------------------------------------------")
print("BİLGİLENDİRME\n")
print("MS VE İY için: EV SAHİBİ KAZANDI -> 1\tBERABERE -> 0\tDEPLASMAN TAKIMI KAZANDI -> 2\n")
print("KG için: İKİ TAKIM GOL ATMIŞSA KG -> 1\t TEK BİR TAKIM ATMIŞSA VEYA İKİ TAKIM DA GOL ATMAMIŞSA -> 0")
print("---------------------------------------------------------------------------------------------\n")
print("SEÇİLEN KARŞILAŞMA BİLGİLERİ\n")
print("GERÇEK MAÇ SONUCU (MS) SKORU: " + ms_score + " ----> " + str(ms_true))
print("TAHMİN EDİLEN MAÇ SONUCU: ------------>", ms_res[0])

print("\nGERÇEK İLK YARI (İY) SONUCU: " + iy_score + " ----> " + str(iy_true))
print("TAHMİN EDİLEN İY SONUCU: ------------->", iy_res[0])

print("\nGERÇEK KARŞILIKLI GOL (KG) SONUCU: -------------> ", kg_true)
print("TAHMİN EDİLEN KG SONUCU: ---------------------->", kg_res[0])
```
